# Check calculations

```python
calc_styles = []
for style in iprPy.recordmanager.loaded_styles:
    if 'calculation_' in style:
        calc_styles.append(style)
calc_styles
```

```python
calc_styles = [
    'calculation_isolated_atom',
    'calculation_diatom_scan',
    'calculation_E_vs_r_scan',
    'calculation_relax_box',
    'calculation_relax_dynamic',
    'calculation_relax_static',
]
```

```python
oldpotid = '2016--Gibson-J--Ti'
newpotid = '2016--Gibson-J-S-Srinivasan-S-G-Baskes-M-I-et-al--Ti'

oldimpid = None#'2021--Mendelev-M--Al-Sm--LAMMPS--ipr1'
newimpid = None#'2021--Song-H--Al-Sm--LAMMPS--ipr1'
```

```python
for calc_style in calc_styles:
    if calc_style in ['calculation_crystal_space_group']:
        continue
    print(calc_style, end=' ')
    records = master.get_records(calc_style, potential_id=oldpotid)
    print(len(records))
    
    for record in records:
        print(' -', record.name)
        if newimpid is not None:
            record.model[record.modelroot]['potential-LAMMPS']['id'] = newimpid
        if newpotid is not None:
            record.model[record.modelroot]['potential-LAMMPS']['potential']['id'] = newpotid
        master.update_record(record)
        iprhub.update_record(record)
        #remote.update_record(record)
    
```

```python
records = master.get_records('calculation_diatom_scan', potential_id=oldpotid)
records
```
